### Data Generation

Realistics synthetic data of 50K rows.

```python
import pandas as pd
import numpy as np

import seaborn as sns
import matplotlib.pyplot as plt

from pathlib import Path
```

```python
np.random.seed(42)

N = 50_000
print(f"Generating {N:,} customer records...")
```

```python
# --- Customer base ---
customer_ids = np.arange(1, N + 1)
signup_dates = pd.date_range("2023-01-01", periods=N, freq="1h")
plan_type = np.random.choice(["free", "basic", "pro"], N, p=[0.50, 0.30, 0.20])

# --- Behavioral features (correlated with churn) ---
# Pro users have longer sessions; free users shorter
base_session = np.where(plan_type == "pro", 300,
               np.where(plan_type == "basic", 180, 90))
session_duration = np.random.exponential(base_session).clip(0, 3600).astype("int64")

# Pages viewed — higher for engaged users
pages_viewed = np.random.poisson(
    np.where(plan_type == "pro", 8, np.where(plan_type == "basic", 5, 3))
).clip(0, 50)

# Support tickets — churners raise more tickets
support_tickets = np.random.poisson(
    np.where(plan_type == "free", 0.8, 0.2)
).clip(0, 10)

# Days inactive — churners go quiet before leaving
days_inactive = np.random.exponential(
    np.where(plan_type == "pro", 5, np.where(plan_type == "basic", 10, 20))
).clip(0, 90).astype(int)

# Login count last 30 days
login_count_30d = np.random.poisson(
    np.where(plan_type == "pro", 20, np.where(plan_type == "basic", 10, 4))
).clip(0, 60)

# Feature used count (product engagement)
features_used = np.random.poisson(
    np.where(plan_type == "pro", 8, np.where(plan_type == "basic", 4, 1))
).clip(0, 20)

# --- Churn label (realistically correlated with features) ---
# Build churn probability from feature signals
churn_score = (
    0.3 * (days_inactive / 90)           # longer inactive → higher churn
    + 0.2 * (support_tickets / 10)        # more tickets → higher churn
    - 0.2 * (session_duration / 3600)     # longer sessions → lower churn
    - 0.15 * (login_count_30d / 60)       # more logins → lower churn
    - 0.1 * (features_used / 20)          # more features used → lower churn
    + 0.05 * (plan_type == "free").astype(float)
    + np.random.normal(0, 0.05, N)        # noise
).clip(0, 1)

churned = (churn_score > np.percentile(churn_score, 93)).astype(int)  # ~7% churn rate

# --- Assemble DataFrame ---
df = pd.DataFrame({
    "customer_id":      customer_ids,
    "signup_date":      signup_dates,
    "plan_type":        plan_type,
    "session_duration_sec": session_duration,
    "pages_viewed":     pages_viewed,
    "support_tickets":  support_tickets,
    "days_inactive":    days_inactive,
    "login_count_30d":  login_count_30d,
    "features_used":    features_used,
    "churned":          churned,
})

# --- Introduce realistic data issues ---
# 3% null rate on session_duration (mimics tracking failures)
null_mask = np.random.rand(N) < 0.03
df.loc[null_mask, "session_duration_sec"] = np.nan

# A few outliers (data entry errors)
df.loc[np.random.choice(N, 10), "session_duration_sec"] = 99999

Path("data").mkdir(exist_ok=True)
df.to_parquet("data/raw_events.parquet", index=False)

print(f"Saved: data/raw_events.parquet")
print(f"\nDataset shape: {df.shape}")
print(f"Churn rate: {churned.mean()*100:.1f}%")
print(f"Null rate (session_duration): {null_mask.mean()*100:.1f}%")
print(f"\nSample:\n{df.head(3).to_string()}")
```

### Feature Engineering Pipeline

Cleans raw data, engineers features, validates quality, saves feature table.

**In production:** this logic runs as a PySpark job on Dataproc, output to BigQuery.
Run: python step2_feature_pipeline.py

```python
import json
```

#### Load raw data

```python
df = pd.read_parquet("data/raw_events.parquet")
print(f"\nLoaded raw data: {df.shape[0]:,} rows, {df.shape[1]} columns")
```

```python
df.head()
```

#### Data Quality Checks (Great Expectations style)

```python
df.isnull().mean()
```

```python
issues = []

null_rates = df.isnull().mean()
for col, rate in null_rates.items():
    if rate > 0.05:
        issues.append(f"  FAIL: {col} null rate = {rate:.1%} (threshold 5%)")
    elif rate > 0:
        print(f"  WARN: {col} null rate = {rate:.1%}")

neg_sessions = (df["session_duration_sec"] < 0).sum()
if neg_sessions > 0:
    issues.append(f"  FAIL: {neg_sessions} negative session_duration values")

if issues:
    for i in issues:
        print(i)
else:
    print("  All checks passed.")
```

#### Fix Outliers & Nulls 

```python
sns.histplot(df["session_duration_sec"], )

plt.xscale("log")
plt.title("Distribution of Session Duration (Sec)")
plt.xlabel("Session Duration (Sec)")

plt.show()
```

```python
# Cap session_duration at 99th percentile (excludes nulls)
p99 = df["session_duration_sec"].quantile(0.99)

no_outliers = (df["session_duration_sec"] > p99).sum() # threshold
df["session_duration_sec"] = df["session_duration_sec"].clip(upper=p99)
print(f"  Capped {no_outliers} outliers at p99={p99:.0f}s")

# Impute nulls with median (fit on full dataset here; in production: fit on train only)
median_session = df["session_duration_sec"].median()
null_count = df["session_duration_sec"].isnull().sum()
df["session_duration_sec"] = df["session_duration_sec"].fillna(median_session)
print(f"  Imputed {null_count} nulls with median={median_session:.0f}s")
```

```python
sns.histplot(df["session_duration_sec"], )

plt.title("Distribution of Session Duration (Sec)")
plt.xlabel("Session Duration (Sec)")

plt.show()
```

#### Feature Engineering 

```python
# Take a very high value in the dataset (top 1% boundary), 
# then take 30% of it to define a lower threshold for filtering, segmentation, or feature engineering.

df["login_count_30d"].quantile(0.99) * 0.3
```

```python
# Engagement score (composite)
df["engagement_score"] = (
    (df["session_duration_sec"] / p99) * 0.3
    + (df["login_count_30d"] / df["login_count_30d"].quantile(0.99)) * 0.3
    + (df["features_used"] / df["features_used"].quantile(0.99)) * 0.2
    + (df["pages_viewed"] / df["pages_viewed"].quantile(0.99)) * 0.2
).clip(0, 1).round(4)

# Risk score (inverse engagement, combined with support friction)
df["risk_score"] = (
    (df["days_inactive"] / 90) * 0.4
    + (df["support_tickets"] / df["support_tickets"].quantile(0.99).clip(1)) * 0.3
    + (1 - df["engagement_score"]) * 0.3
).clip(0, 1).round(4)

# Recency bucket (categorical)
df["recency_bucket"] = pd.cut(
    df["days_inactive"],
    bins=[0, 7, 14, 30, 90],
    labels=["active", "at_risk", "dormant", "churning"],
    right=True
).astype(str)

# Ticket rate per login (avoid division by zero)
df["tickets_per_login"] = (
    df["support_tickets"] / (df["login_count_30d"] + 1)
).round(4)

print(f"  Created: engagement_score, risk_score, recency_bucket, tickets_per_login")
```

#### Build Final Feature Table

```python
df.columns
```

```python
FEATURE_COLS = df.columns

features = df[FEATURE_COLS].copy()
features["feature_created_at"] = pd.Timestamp("2026-05-22")  # fixed reference date

Path("features").mkdir(exist_ok=True)
features.to_parquet("features/feature_table.parquet", index=False)
```

#### Summary Stats

```python
print(f"\n  Churn rate: {features['churned'].mean()*100:.1f}%")
print(f"  Positive examples: {features['churned'].sum():,}")

print("\n  Feature stats by churn label:")
summary = features.groupby("churned")[
    ["session_duration_sec", "days_inactive", "support_tickets",
     "engagement_score", "risk_score"]
].mean().round(2)
print(summary.to_string())

print(f"\n  Recency bucket distribution:")
print(features.groupby("recency_bucket")["churned"].agg(["count","mean"]).round(3).to_string())

# Save quality report
quality_report = {
    "rows": len(features),
    "churn_rate": float(features["churned"].mean()),
    "null_rates": features.isnull().mean().round(4).to_dict(),
    "outliers_capped": int(no_outliers),
    "nulls_imputed": int(null_count),
}
with open("features/quality_report.json", "w") as f:
    json.dump(quality_report, f, indent=2)

print(f"\nSaved: features/feature_table.parquet")
print(f"Saved: features/quality_report.json")
print("\nPipeline complete.")
```
